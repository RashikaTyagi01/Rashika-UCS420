Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt
M=float(input("enter a value for M: "))
x = np.linspace(-10, 10, 100)
y1 = M * x**2
y2 = M * np.sin(x)
plt.plot(x,y1,'r--', label='y = M*x^2')
plt.plot(x,y2,'b-',label='y = M*sin(x)')
plt.legend()
plt.grid()
plt.title('Mathematical Functions')
plt.show()



```

Q2. Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = {'Subject' : ['Maths','English','Biology','Chemistry','Physics'],
        'Score' : [95, 84, 87, 90, 93]
}
df = pd.DataFrame(data)
print(df)
sns.set_theme(style='darkgrid')
ax = sns.barplot(x='Subject', y='Score',data = df, palette='muted')
for p in ax.patches:
  ax.annotate(f"{p.get_height()}",
              (p.get_x() + p.get_width() / 2., p.get_height()), ha = 'center', va = 'bottom',fontsize=12, fontweight='bold', color='black')
plt.title('Scores in Different Subjects',fontweight='bold')
plt.xlabel('Subjects', fontsize=14)
plt.ylabel('Scores',fontsize=14)
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter roll number as seed: "))
np.random.seed(roll_number)
data = np.random.randn(50)
print(data)
fig, ax = plt.subplots(2,2, figsize=(12, 10))
ax[0,0].plot(np.cumsum(data))
ax[0,0].set_title('Cumulative Sum Plot')
ax[0,0].set_xlabel('Index')
ax[0,0].set_ylabel('Cumulative sum')
ax[0,0].grid(True)

ax[0,1].scatter(range(50), data)
ax[0,1].set_title('Scatter Plot with Noise')
ax[0,1].set_xlabel('Index')
ax[0,1].set_ylabel('Value')
ax[0,1].grid(True)

ax[1, 0].axis('off')
ax[1, 1].axis('off')
plt.tight_layout()
plt.show()
```

Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python

from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['company_sales_data.csv']))
print(df)
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
sns.set_theme(style='darkgrid')

plt.figure(figsize=(10, 6))
sns.lineplot(x='month_number', y='total_profit', data=df, marker='o', color='b')
plt.title('Total Profit Over Months')
plt.xlabel('Months')
plt.ylabel('Total Profit')
plt.grid(True)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
for column in ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']:
  sns.lineplot(x='month_number',y=column, data=df, label=column)
plt.title('Product Sales Over Months')
plt.xlabel('Month')
plt.ylabel('Product Sales')
plt.legend(title='Products')
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 8))
df.drop('month_number', axis=1).sum().plot(kind='bar', color=sns.color_palette('husl'))
plt.title('Total Sales and Profit of All Products')
plt.xlabel('Features/Attributes')
plt.ylabel('Total Value')
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

```
