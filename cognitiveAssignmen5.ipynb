Q1. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find
i. Sum of all elements
ii. Sum of all elements row-wise
iii. Sum of all elements column-wise

```python
import numpy as np
gfg = np.matrix('[4,1,9; 12,3,1; 4,5,6]')

#(i)
print("Sum of all elements: ",np.sum(gfg))
#(ii)
print("Sum of all elements row-wise: ",np.sum(gfg,axis=0))
#(iii)
print("Sum of all elements column-wise: ",np.sum(gfg,axis=1))
```

Q2.a (a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find
i. Sorted array
ii. Indices of sorted array
iii. 4 smallest elements
iv. 5 largest elements

```python
import numpy as np
array = np.array([10, 52, 62, 16, 16, 54, 453])
#(i)
print("Sorted array: ",np.sort(array))
#(ii)
print("Indices of sorted array: ",np.argsort(array))
#(iii)
sorted = np.sort(array)
print("4 smallest elements: ",sorted[:4])
#(iv)
print("5 largest elements: ",sorted[-5:])
```

Q2.b For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find
i. Integer elements only
ii. Float elements only


```python
import numpy as np
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
#(i)
print("The integer elements: ",array[array==array.astype(int)])
#(ii)
print("Float elements: ",array[array == array.astype(float)])

float_elements = array[array != array.astype(int)]
print("float elements: ", float_elements)
```

Q3.a You are given a weekly sales dataset and need to perform various operations
using NumPy broadcasting.
a) Generate your unique sales dataset:
 Take the sum of the ASCII values of the initials of your first and last
name. Call this value X. (If your initials are A B → ASCII sum = 65
+ 66 = 131 → sales = [131, 181, 231, 281, 331].)
 Create a NumPy array sales with values [X, X+50, X+100, X+150,
X+200].

```python
import numpy as np
a = 'R'
b = 'T'
X = ord(a) + ord(b)
print(X)
sales = np.array([X, X+50, X+100, X+150, X+200])
print(sales)
```

Q3.b Compute your personalized tax rate as ((X % 5) + 5) / 100.
 Use broadcasting to apply this tax rate to each sales value.

```python
taxRate = ((X % 5) + 5) / 100
print("After applying tax rate: ")
afterTax = sales*taxRate
print(afterTax)
```

Q3.c Adjust sales based on discount:
 If sales < X+100, apply a 5% discount.
 If sales >= X+100, apply a 10% discount.

```python
afterDiscount = np.where(sales<X+100, sales*0.95, sales*0.90)
print("sales after applying discount: ",afterDiscount)
```

Q3.d Expand sales data for multiple weeks:
 Create a 3×5 matrix representing three weeks of sales by stacking
sales three times using broadcasting.
 Increase sales by 2% per week using element-wise broadcasting.

```python
salesMatrix = np.tile(sales,(3,1))
print("three weeks of sales: ",salesMatrix)

weeklyIncrease = np.array([1.00, 1.02, 1.04]).reshape(3, 1)
adjusted_weekly_sales = salesMatrix * weeklyIncrease
print("weekly sales after 2% increase per week: ",adjusted_weekly_sales)

```

Q4. Generate x values using np.linspace() from -10 to 10 with 100 points. Use
each function from the list below and compute y values using NumPy:
 Y = x2
 Y = sin(x)
 Y = ex
 Y = log(|x| + 1)
Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=np.linspace(-10,10,100)
y1=x**2
plt.plot(x,y1)
plt.title('y=x^2')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
```

```python
y2=np.sin(x)
plt.plot(x,y2)
plt.title('y=sin(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
```

```python
y3=np.exp(x)
plt.plot(x,y3)
plt.title('y=exp(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
```

```python
y4=np.log(np.abs(x)+1)
plt.plot(x,y4)
plt.title('y=log(|x| + 1)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
```
